```python
from tensorflow.keras import models, layers
import numpy as np
import tensorflow as tf
np.random.seed(3)
tf.random.set_seed(3)
dataset = np.loadtxt('data/pima-indians-diabetes.csv', delimiter =',')
x = dataset[:, 0:8]
y = dataset[:, 8]
```

```python
model_name = 'pima-indians-diabetes'
BATCH_SIZE = 10
params = {'model_name':model_name,
          'input_dimension': 8,
          'hidden_dimension1': 12,
          'hidden_dimension2': 8,
          'output_dimension' : 1}
```

```python
class ThoraricSurgery(tf.keras.Model):
    def __init__(self, **kargs): #딕셔너리의 파라미터를 받을 때에는 **을 붙혀야 한다.
        super(ThoraricSurgery, self).__init__(name=kargs['model_name'])
        self.fc1 = layers.Dense(kargs['hidden_dimension1'], input_dim = kargs['input_dimension'], activation = 'relu')
        self.fc2 = layers.Dense(kargs['hidden_dimension2'], input_dim = kargs['input_dimension'], activation = 'relu')
        self.fc3 = layers.Dense(kargs['output_dimension'], activation = 'sigmoid')
        
        
        
    def call(self, x):
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.fc3(x)
        
        return x
```

```python
model = ThoraricSurgery(**params)
model.build((BATCH_SIZE, params['input_dimension']))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics = ['accuracy'])
model.fit(x, y, epochs=200, batch_size=10)
```
