```python
from tensorflow.keras import models, layers
import numpy as np
import tensorflow as tf

np.random.seed(3)
tf.random.set_seed(3)

data_set = np.loadtxt('data/ThoraricSurgery.csv', delimiter=',')

X = data_set[:,0:17]
Y = data_set[:, 17]
```

```python
model_name = 'thoraric_surgery'
BATCH_SIZE = 10
params = {'model_name':model_name,
          'input_dimension': 17,
          'hidden_dimension': 30,
          'output_dimension' : 1}
```

```python
class ThoraricSurgery(tf.keras.Model):
    def __init__(self, **kargs): #딕셔너리의 파라미터를 받을 때에는 **을 붙혀야 한다.
        super(ThoraricSurgery, self).__init__(name=kargs['model_name']) #처음부터 여기까지는 고정해서 작성한다고 보면 됨
        self.fc1 = layers.Dense(kargs['hidden_dimension'], input_dim = kargs['input_dimension'], activation = 'relu')
        self.fc2 = layers.Dense(kargs['output_dimension'], activation = 'sigmoid')
    
    def call(self, x):
        x = self.fc1(x)
        x = self.fc2(x)
        return x
```

```python
model = ThoraricSurgery(**params)
```

```python
# model.summary() 아래 코드를 build를 안한 상태로 돌리면 오류가 뜬다. 그러므로 아래와같이 빌드를 해서 돌리면 오류가 뜨지 않는다.
model.build((BATCH_SIZE, params['input_dimension']))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics = ['accuracy'])
model.fit(X, Y, epochs=100, batch_size=10)
```
